```julia
using RDatasets

iris = dataset("datasets", "iris")
iris
```

```julia
iris[80, "Species"]
```

```julia
collect(combinations(1:4, 2))
```

```julia
using Combinatorics
using Plots

# This displays a scatter plot of sepal length vs sepal width
scatter(
    iris.SepalLength,
    iris.SepalWidth,
    xlabel="Sepal Length",
    ylabel="Sepal Width",
    legend=false,
)
```

```julia
pairs = collect(combinations(1:4, 2)) # Get pairs of indices from 1 to 4

pairs
```

```julia
for pair in pairs
    df = iris[!, pair] # Get only columns with indices in the pairs
    xlabel, ylabel = names(iris)[pair] # Get associated column names

    display(
        scatter(
            df[!, 1], # Get the column at the first index in the pair
            df[!, 2], # Get the column at the second index in the pair
            xlabel=xlabel,
            ylabel=ylabel,
            legend=false
        )
    )
end
```
